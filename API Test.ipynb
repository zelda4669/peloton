```python
import config
import requests
import pandas as pd
```

```python
conn = requests.Session()
payload = {'username_or_email': config.email, 'password':config.password}
conn.post('https://api.onepeloton.com/auth/login', json=payload)
```

```python
rides = conn.get('https://api.onepeloton.com/api/v2/ride/archived?browse_category=')
```

```python
rides = rides.json()
```
